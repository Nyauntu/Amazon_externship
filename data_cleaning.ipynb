```python
df = df.drop(columns=['company_name','flag_covid','review_advice','company_name','rating_compensation_benefits','rating_career_opportunities','rating_senior_leadership','rating_diversity_inclusion','rating_culture_values','rating_culture_values','count_helpful','review_advice','career_opportunities_rating','count_unhelpful'])  # Dropping irrelevant or not so much useful columns'
```

```python
print(df.head())
```

```python
df.drop(columns=['rating_date','flags_ceo_approval','flags_recommend_frend','rating_work_life','flag_featured','flags_business_outlook','employee_length'])
```

```python
print(df.head())
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
print(df.head())
```

```python
df.drop(columns=[
    'rating_date',
    'flags_ceo_approval',
    'flags_recommend_frend',
    'rating_work_life',
    'flag_featured',
    'flags_business_outlook'
], inplace=True)
```

```python
print(df.head())
```

```python
df = df.dropna(subset=['summary'])
```

```python
df = df.drop_duplicates()
```
