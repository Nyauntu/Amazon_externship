```python
import pandas as pd

# Load your dataset
df = pd.read_csv('/content/combined_sentiment_data.csv')
```

```python
#now in this step, we allocate particular words to themes
theme_keywords = {
    'Burnout/Physical strain': ['tired', 'exhausted', 'overworked', 'no break', 'stress','hard','hours','pressure','long shifts','difficult''heavy','hours'],
    'Career growth': ['training', 'management', 'no support', 'monotonous', 'boring','boredom','promote','promotion'],
    'Managerial issue': ['favoritism', 'schedule', 'promotion', 'number'],

}
```

```python
def tag_themes(text, theme_keywords):
    matched_themes = []
    for theme, keywords in theme_keywords.items():
        if any(keyword in str(text).lower() for keyword in keywords):
            matched_themes.append(theme)
    return ', '.join(matched_themes) if matched_themes else ''
```

```python
df['Theme'] = df['cleaned_text'].apply(lambda x: tag_themes(x, theme_keywords))
```

```python
df.to_csv('combined_with_themes.csv', index=False)
```
